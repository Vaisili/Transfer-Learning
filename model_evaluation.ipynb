```python
import torch
import numpy as np
from torchvision import datasets, transforms
from model import ResNetFashion
from data_loader import FashionDataset


#function to compute topk accuracy
def compute_acc(output, target, top=1):
    with torch.no_grad():
        batch_size = target.size(0)

        _, pred = output.topk(top, 1, largest=True, sorted=True)
        pred = pred.t()
        correct = pred.eq(target.view(1, -1).expand_as(pred))

        correct_k = correct[:top].view(-1).float().sum(0, keepdim=True)
        acc = correct_k.mul_(100.0 / batch_size)
        return acc
```
